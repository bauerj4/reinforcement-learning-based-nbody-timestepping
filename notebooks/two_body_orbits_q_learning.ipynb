# Orbit Energies for 2-Body Problem

This notebook generates plots for my study of Q-Learning applied to the 2-body problem. 

## Baseline: Fixed Timestep for Integration

Make plots to establish a baseline for the problem.

```python
import pandas as pd
import os

# Data directories
data_dir = "../data/"

# Circular orbit
euler_circular_baseline = "big_body_small_body_circular_euler_fixed_timestep" 
leapfrog_order_1_circular_baseline = "big_body_small_body_circular_leapfrog_1_fixed_timestep" 
leapfrog_order_2_circular_baseline = "big_body_small_body_circular_leapfrog_2_fixed_timestep"

# e = 0.95 orbit
euler_elliptical_baseline = "big_body_small_body_e_0.95_euler_fixed_timestep" 
leapfrog_order_1_elliptical_baseline = "big_body_small_body_e_0.95_leapfrog_1_fixed_timestep" 
leapfrog_order_2_elliptical_baseline = "big_body_small_body_e_0.95_leapfrog_2_fixed_timestep"

# Load energies

euler_circular_baseline_energies = pd.read_csv(
    os.path.join(data_dir, euler_circular_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

leapfrog_order_1_circular_baseline_energies = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_1_circular_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

leapfrog_order_2_circular_baseline_energies = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_2_circular_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

# Load particles
euler_circular_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, euler_circular_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)

leapfrog_order_1_circular_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_1_circular_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)

leapfrog_order_2_circular_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_2_circular_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)


euler_elliptical_baseline_energies = pd.read_csv(
    os.path.join(data_dir, euler_elliptical_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

leapfrog_order_1_elliptical_baseline_energies = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_1_elliptical_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

leapfrog_order_2_elliptical_baseline_energies = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_2_elliptical_baseline, "0.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)

# Load particles
euler_elliptical_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, euler_elliptical_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)

leapfrog_order_1_elliptical_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_1_elliptical_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)

leapfrog_order_2_elliptical_baseline_p1 = pd.read_csv(
    os.path.join(data_dir, leapfrog_order_2_elliptical_baseline, "episode.particle.0.0.dat"),
    names=["Time (Gyr)", "x", "y", "z", "vx", "vy", "vz", "ax", "ay", "az"]
)
```

```python
import matplotlib.pylab as plt
import numpy as np

# T = 2 pi sqrt(r**3 / GM)
euler_circular_relative_energies = abs(euler_circular_baseline_energies["Energy (G=1 units)"] - euler_circular_baseline_energies["Energy (G=1 units)"].iloc[0]) / -euler_circular_baseline_energies["Energy (G=1 units)"].iloc[0]
leapfrog_order_1_circular_relative_energies = abs(leapfrog_order_1_circular_baseline_energies["Energy (G=1 units)"] - leapfrog_order_1_circular_baseline_energies["Energy (G=1 units)"].iloc[0]) / -leapfrog_order_1_circular_baseline_energies["Energy (G=1 units)"].iloc[0]
leapfrog_order_2_circular_relative_energies = abs(leapfrog_order_2_circular_baseline_energies["Energy (G=1 units)"] - leapfrog_order_2_circular_baseline_energies["Energy (G=1 units)"].iloc[0]) / -leapfrog_order_2_circular_baseline_energies["Energy (G=1 units)"].iloc[0]


euler_elliptical_relative_energies = abs(euler_elliptical_baseline_energies["Energy (G=1 units)"] - euler_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]) / -euler_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]
leapfrog_order_1_elliptical_relative_energies = abs(leapfrog_order_1_elliptical_baseline_energies["Energy (G=1 units)"] - leapfrog_order_1_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]) / -leapfrog_order_1_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]
leapfrog_order_2_elliptical_relative_energies = abs(leapfrog_order_2_elliptical_baseline_energies["Energy (G=1 units)"] - leapfrog_order_2_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]) / -leapfrog_order_2_elliptical_baseline_energies["Energy (G=1 units)"].iloc[0]

# Create a figure with two subplots sharing the y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Plot the first orbit on the first panel
ax1.plot(euler_circular_baseline_energies["Time (Gyr)"], euler_circular_relative_energies, label="Forward Euler")
ax1.plot(leapfrog_order_1_circular_baseline_energies["Time (Gyr)"], leapfrog_order_1_circular_relative_energies, label="Implicit Euler")
ax1.plot(leapfrog_order_2_circular_baseline_energies["Time (Gyr)"], leapfrog_order_2_circular_relative_energies, label="Verlet")
ax1.set_xlabel("$t$ (Gyr)")
ax1.set_ylabel(r"$\left\vert\delta E / E \right\vert$")
ax1.set_title("Circlular Orbit")
ax1.grid(True)

# Plot the second orbit on the second panel
ax2.plot(euler_elliptical_baseline_energies["Time (Gyr)"], euler_elliptical_relative_energies, label="Forward Euler")
ax2.plot(leapfrog_order_1_elliptical_baseline_energies["Time (Gyr)"], leapfrog_order_1_elliptical_relative_energies, label="Implicit Euler")
ax2.plot(leapfrog_order_2_elliptical_baseline_energies["Time (Gyr)"], leapfrog_order_2_elliptical_relative_energies, label="Verlet")
ax2.set_xlabel("$t$ (Gyr)")
ax2.set_title("$e=0.95$ Elliptical Orbit")
ax2.legend()
ax2.grid(True)

# Display the plot
plt.yscale("log")
plt.tight_layout()
plt.show()
# plt.savefig("../data/figures/two-body-orbital-energies.png")
```

Do reinforcement learning and example the Q tables.

```python
from scipy.interpolate import interp1d

agent_elliptical = "big_body_small_body_e_0.95_agent_learned_timestep"

agent_elliptical_energies = pd.read_csv(
    os.path.join(data_dir, agent_elliptical, "99.energies.dat"),
    names=["Time (Gyr)", "Min Timestep (Gyr)", "Max Timestep (Gyr)", "Energy (G=1 units)", "Acceleration calculations"]
)
agent_elliptical_relative_energies = abs(agent_elliptical_energies["Energy (G=1 units)"] - agent_elliptical_energies["Energy (G=1 units)"][0]) / -agent_elliptical_energies["Energy (G=1 units)"][0]

fixed_tstep_relative_energy = interp1d(leapfrog_order_2_circular_baseline_energies["Time (Gyr)"], leapfrog_order_2_elliptical_relative_energies, fill_value="extrapolate")
# Create a figure with two subplots sharing the y-axis
fig, ax1 = plt.subplots(1, 1, figsize=(6, 6), sharey=True)

# Plot the first orbit on the first panel
ax1.plot(agent_elliptical_energies["Time (Gyr)"], agent_elliptical_relative_energies / fixed_tstep_relative_energy(agent_elliptical_energies["Time (Gyr)"]), label="Verlet")
#ax1.plot(leapfrog_order_2_elliptical_baseline_energies["Time (Gyr)"], leapfrog_order_2_elliptical_relative_energies, label="Verlet")
ax1.set_xlabel("$x$ (kpc / h)")
ax1.set_ylabel(r"$\left\vert\delta E / E \right\vert$")
ax1.set_title("Circlular Orbit")
ax1.grid(True)
plt.xlim(0, np.amax(agent_elliptical_energies["Time (Gyr)"],))
# plt.ylim(7e-4, 1.5e-3)
plt.yscale("log")
```
